```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
```

```python
from langchain_core.documents import Document

doc1 = Document(
    page_content="Virat Kohli is a former captain of the Indian cricket team and one of the greatest batsmen in the world.",
    metadata={
        "name": "Virat Kohli",
        "country": "India",
        "role": "Batsman"
    }
)

doc2 = Document(
    page_content="MS Dhoni is known for his calm leadership and finishing skills. He led India to multiple ICC trophies.",
    metadata={
        "name": "MS Dhoni",
        "country": "India",
        "role": "Wicketkeeper Batsman"
    }
)

doc3 = Document(
    page_content="Sachin Tendulkar is called the God of Cricket and holds the record for most international runs.",
    metadata={
        "name": "Sachin Tendulkar",
        "country": "India",
        "role": "Batsman"
    }
)

doc4 = Document(
    page_content="Babar Azam is one of the top modern-day batsmen and captain of Pakistan cricket team.",
    metadata={
        "name": "Babar Azam",
        "country": "Pakistan",
        "role": "Batsman"
    }
)

doc5 = Document(
    page_content="Joe Root is an English cricketer and a key player in Test cricket known for his consistency.",
    metadata={
        "name": "Joe Root",
        "country": "England",
        "role": "Batsman"
    }
)
```

```python
docs = [doc1,doc2,doc3,doc4,doc5]
```

```python
vector_store = Chroma(embedding_function=HuggingFaceEmbeddings(),
                      persist_directory='chroma_db',
                      collection_name='sample')
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include=['embeddings','documents','metadatas'])
```

```python
vector_store.similarity_search(query='who among these are a Wicketkeeper ',
                               k=2)
```

```python
vector_store.similarity_search_with_score(query='who among these are a wicketkeeper',
                                          k=2)
```

```python
vector_store.similarity_search_with_score(query='',
                                          filter={'country': 'India'})
```

```python
# update documents
updated_doc1 = Document(
page_content="Virat Kohli, the former captain of Royal Challengers Bangalore (RCB), is renowned for his aggressive leadership and consister",
metadata={"country": "Royal Challengers Bangalore"}
)
vector_store.update_document (document_id='8b40751b-dc1b-4447-8229-b453a84e6bae', document=updated_doc1)
```
